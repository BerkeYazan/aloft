```python
import pandas as pd

# Load the enriched dataset
df = pd.read_csv('Data/quotes_with_author_language.csv')

# Filter for authors that weren't found
not_found_authors = df[df['language_or_country'] == 'Not Found']['AUTHOR'].unique()

print(f"Total unique authors that were not found: {len(not_found_authors)}")
print(f"This represents {len(not_found_authors)/df['AUTHOR'].nunique():.2%} of all unique authors\n")

print("--- Complete List of Authors Not Found ---")
for i, author in enumerate(sorted(not_found_authors), 1):
    print(f"{i:3d}. {author}")
```
