```python
# ═══════════ Notebook  :  Surprisal  – CELL 1 ═══════════

from pathlib import Path
import torch
from transformers import GPT2TokenizerFast, GPT2LMHeadModel

# --- paths ---------------------------------------------------------
NB_DIR     = Path.cwd()                               # notebook folder
MODEL_NAME = "gpt2-medium"                            # 355 M params
DEVICE     = "mps" if torch.backends.mps.is_available() else "cpu"
print("Device →", DEVICE)

# --- load ----------------------------------------------------------
tokenizer = GPT2TokenizerFast.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token
model     = GPT2LMHeadModel.from_pretrained(MODEL_NAME).to(DEVICE).eval()
torch.manual_seed(42)

print("✓ GPT-2-medium loaded (tokeniser + model)")
```

```python
# ═══════════ Notebook : Surprisal – CELL 2 ═══════════
import math, numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
from tqdm.auto import tqdm
from statsmodels.stats.multitest import multipletests
from scipy.stats import mannwhitneyu

# --- 1. paths & pilot toggle ---------------------------------------
DATA_CSV   = Path.cwd() / "ALOFT.csv"
OUT_DIR    = Path.cwd() / "features"
OUT_DIR.mkdir(exist_ok=True)

SAMPLE_FRAC = 1.0          # 0.10 pilot, 1.0 full

df = pd.read_csv(DATA_CSV).reset_index(drop=True)

# --- 2. columns ----------------------------------------------------
COLUMNS = {
    "Goodreads Sample Quote"             : "surp_sample",
    "Goodreads Popular Quote"            : "surp_popular",
    "Google Books Length Matched Snippet": "surp_snippet",
    "T50 Quote"                          : "surp_t50",
    "Non-Literary Baseline"              : "surp_nonlit",
}

MIN_TOKENS, BATCH = 10, 8

# --- 3. helper -----------------------------------------------------
def bits_per_token(ids):
    with torch.no_grad():
        loss = model(ids.to(DEVICE), labels=ids.to(DEVICE)).loss
    return (loss.item() * ids.size(1) / math.log(2)) / ids.size(1)

def mean_surprisal(texts):
    enc = tokenizer(texts, return_tensors="pt", padding=True,
                    add_special_tokens=False).input_ids
    return [bits_per_token(enc[i:i+1]) for i in range(enc.size(0))]

# --- 4. compute surprisal -----------------------------------------
scores = pd.DataFrame({"row_id": df.index})

for src, out_col in COLUMNS.items():
    print(src)
    texts, col_vals = df[src].fillna("").tolist(), []
    for i in tqdm(range(0, len(texts), BATCH)):
        chunk = texts[i:i+BATCH]
        keep  = [len(tokenizer.tokenize(t)) >= MIN_TOKENS for t in chunk]
        vals  = mean_surprisal([t for t,k in zip(chunk,keep) if k]) if any(keep) else []
        ptr = 0
        for k in keep:
            col_vals.append(vals[ptr] if k else np.nan)
            if k: ptr += 1
    scores[out_col] = col_vals

scores.to_csv(OUT_DIR / "surprisal_scores.csv", index=False)
print("✓ saved features/surprisal_scores.csv", scores.shape)

# ────────────────── 5. visualisation & stats ───────────────────────
palette = {
    "Goodreads Sample Quote"             : COLOR_SAMPLE_QUOTE,
    "Goodreads Popular Quote"            : COLOR_POPULAR_QUOTE,
    "Google Books Length Matched Snippet": COLOR_MATCHED_SNIPPET,
    "T50 Quote"                          : COLOR_T50_QUOTE,
    "Non-Literary Baseline"              : COLOR_NONLIT_BASELINE,
}

# long DF for plotting
long = scores.melt(id_vars="row_id", var_name="Metric", value_name="Bits").dropna()
long["TextType"] = long["Metric"].map({v:k for k,v in COLUMNS.items()})

plt.figure(figsize=(12,6))
for src,c in palette.items():
    sns.kdeplot(long.loc[long.TextType==src,"Bits"], label=src, fill=True,
                alpha=.55, linewidth=0, color=c)
plt.title("Mean Sub-token Surprisal (GPT-2 medium)", weight="bold")
plt.xlabel("Bits per sub-token  – higher = more surprising"); plt.ylabel("Density")
plt.legend(fontsize=8, loc="upper left"); sns.despine(); plt.tight_layout(); plt.show()

# descriptive stats
desc = (long.groupby("TextType")["Bits"]
              .describe()[["count","mean","std","25%","50%","75%"]].round(3))
print("\n=== Descriptive statistics (bits per sub-token) ===")
print(desc)

# Mann-Whitney vs snippet baseline
baseline = "Google Books Length Matched Snippet"
base_vals = long.loc[long.TextType==baseline,"Bits"].dropna()

def cliffs(a,b):
    m,n=len(a),len(b)
    gt=sum(x>y for x in a for y in b)
    lt=sum(x<y for x in a for y in b)
    return (gt-lt)/(m*n)

rows,pvals=[],[]
for src in palette:
    if src==baseline: continue
    vals = long.loc[long.TextType==src,"Bits"].dropna()
    u,p = mannwhitneyu(vals, base_vals, alternative="greater")
    rows.append({"QuoteType":src,"U":u,"p_raw":p,"CliffΔ":cliffs(vals.values, base_vals.values)})
    pvals.append(p)

_,q,_,_=multipletests(pvals, method="fdr_bh")
for r,qv in zip(rows,q): r["p_FDR"]=qv

fmt={"p_raw":lambda x:f"{x:.2e}","p_FDR":lambda x:f"{x:.2e}","CliffΔ":lambda x:f"{x:.3f}"}
print("\n=== Mann-Whitney U (one-sided: quote > snippet) ===")
print(pd.DataFrame(rows).to_string(index=False, formatters=fmt))
```

```python
# ═══════════ Notebook : Surprisal – CELL 3 (Top-50 full text) ═══════════
import pandas as pd
pd.set_option("display.max_colwidth", None)   # show entire passage

# 1. merge original text with surprisal scores
text_cols = list(COLUMNS.keys())                       # the five passage columns
merged = pd.concat(
    [df.reset_index(drop=True)[text_cols],              # raw texts
     scores.drop(columns="row_id").reset_index(drop=True)], axis=1
)

# 2. long-format table: one row per passage per text type
rows = []
for src, col in COLUMNS.items():
    tmp = merged[[src, col]].copy().rename(columns={src: "Text", col: "Surprisal"})
    tmp["TextType"] = src
    rows.append(tmp)

long_top = pd.concat(rows).dropna(subset=["Surprisal"])

# 3. global Top-50
top50 = (long_top.sort_values("Surprisal", ascending=False)
                   .head(50)
                   .reset_index(drop=True))   # keep full text

print("=== Top 50 passages by mean surprisal (bits per sub-token) ===")
display(top50[["TextType", "Surprisal", "Text"]])
```
